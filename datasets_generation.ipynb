```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
bov   = pd.read_csv('data/raw/ibov.csv')
dol   = pd.read_csv('data/raw/dolar_total.csv') 
dj    = pd.read_csv('data/raw/dj.csv')
selic = pd.read_csv('data/raw/taxa_selic.csv') 
vale  = pd.read_csv('data/raw/vale_full.csv') 
```

```python
opcao1 = pd.read_csv('data/processed/sentiment_op1.csv')
opcao1.rename(columns = {'Data': 'data'}, inplace = True)
```

```python
opcao2 = pd.read_csv('data/processed/sentiment_op2.csv')
opcao2.rename(columns = {'Data': 'data'}, inplace = True)
```

```python
opcao3 = pd.read_csv('data/processed/sentiment_op3.csv')
opcao3.rename(columns = {'Data': 'data'}, inplace = True)
```

### Vale + Selic

```python
vale_selic = vale.merge(selic, on = 'data')
vale_selic.rename(columns = {'valor': 'selic'}, inplace = True)
vale_selic.to_csv('data/processed/vale_selic.csv', index = False)
vale_selic.head()
```

### Vale + Ibov

```python
vale_bov = vale.merge(bov[['data', 'Último']], on = 'data')
vale_bov.rename(columns={'Último': 'ibov'}, inplace=True)
vale_bov.to_csv('data/processed/vale_bov.csv', index = False)
vale_bov.head()
```

### Vale + Ibov + DJ

```python
vale_bov_dj = vale_bov.merge(dj, on='data')
vale_bov_dj.to_csv('data/processed/vale_bov_dj.csv', index = False)
vale_bov_dj.head()
```

### Vale + Ibov + Dolar

```python
vale_bov_dol = vale_bov.merge(dol, on='data', how = 'left')
vale_bov_dol.rename(columns={'abertura': 'aber_dol', 'fechamento': 'fech_dol'}, inplace=True)
vale_bov_dol.to_csv('data/processed/vale_bov_dol.csv', index=False)
vale_bov_dol.head()
```

### Vale + Ibov + Selic

```python
vale_bov.selic = vale_bov.merge(selic, on = 'data')
vale_bov.selic.rename(columns = {'valor': 'selic'}, inplace = True)
vale_bov.selic['selic'] = vale_bov.selic['selic'] * 100
vale_bov.selic.to_csv('data/processed/vale_bov_sel.csv', index=False)
vale_bov.selic.head()
```

### Vale + Ibov + Sentimento op1

```python
vale_bov_opcao1_full = vale_bov.merge(opcao1[['data', 'Sentiment', 'qtde']], on = 'data', how = 'left')
vale_bov_opcao1_full.fillna(0, inplace=True)
vale_bov_opcao1_full.to_csv('data/processed/vale_bov_opcao1.csv', index = False)
vale_bov_opcao1_full.head()
```

### Vale + Ibov + Sentimento op2

```python
vale_bov_opcao2_full = vale_bov.merge(opcao2.drop(columns=['Polaridade', 'Subjetividade']), on = 'data', how = 'left')
vale_bov_opcao2_full.fillna(0, inplace=True)
vale_bov_opcao2_full.to_csv('data/processed/vale_bov_opcao2.csv', index = False)
vale_bov_opcao2_full.head()
```

### Vale + Ibov + Sentimento op3

```python
vale_bov_opcao3_full = vale_bov.merge(opcao3, on = 'data', how = 'left')
vale_bov_opcao3_full.fillna(0, inplace=True)
vale_bov_opcao3_full.to_csv('data/processed/vale_bov_opcao3.csv', index = False)
vale_bov_opcao3_full.head()
```
